## Getting In and Getting Out

Once you have the Anaconda python package installed, in your user folder there shold be an `anaconda3` directory. 

### Openning Jupyter Notebooks

#### 1. Open the Anaconda-Navigator Desktop software

- Open `anaconda3` directory in your user folder.
- Double-click the `Anaconda-Navigator` program



#### 2.1 `Launch` JupyterLab

- To launch the JupyterLab software, click the _launch_ button in the Anaconda-Navigator interface (shown above)
- Using the file browser in JupyterLab find the notebook you want to open and double-click on the icon.

<img src='../images/jupyterlab.png'>


__NOTE:__ 
- You can also launch `JupyterLab` by typing, in the `Terminal` or `Anaconda Prompt`,:

     $ jupyterlab

#### 2.2 `Launch` Jupyter Notebook

- To launch the Jupyter Notebook software, click the _launch_ button in the Anaconda-Navigator interface (shown above)
- A Terminal windows will open and the default web browser window will also open displying the Jupyter Dashboard (shown above)

<img src='../images/jupyterlab.png'>

__NOTE__
- It is __NOT__ possible to double click a Jupyter Notebook to open it.
- The `Terminal` window openned by the Jupyter notebook software __MUST__ remain open while using the Jupyter Notebook.
- You _can_ download the [`Pineapple`](https://nwhitehead.github.io/pineapple/) program  that enables double-click launch of Jupyter Notebooks. The [`nteract`](https://nteract.io) project can also be used to work with Jupyter Notebooks. Both are MacOS only

#### 3. From the command line

Another way to launch the Jupyter notebook is to do it from the command line by typing:

    jupyter notebook

The notebook will open in your default browser. You can also specify which browser to use:

    jupyter notebook --browser=chrome

By default, the notebook server starts on port 8888, or, if this port is not available, to the next available one, such as 9999. You may also specify a port manually:

    jupyter notebook --port 9999

__Browser Compatibility__

The Jupyter Notebook is officially supported by the latest stable versions of the following browsers:
- Chrome
- Safari
- Firefox

The is mainly due to the notebook’s usage of WebSockets and the flexible box model.

### To close a Jupyter Notebook

- From the (Jupyter) _File_ menu select close and halt on any open notebooks
    - Or from the _Running_ tab in the Jupyter Deshboard click the running tab and shutdown any running notebooks.
- Close the Jupyter Dashboard (or quit your browser)
- Close the Terminal window hosting the Jupyter server or you can kill Jupyter by hitting “Ctrl + C” in the terminal window
